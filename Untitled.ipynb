```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
```

```python
import codeless
```

```python
df = pd.read_csv('AdvanceHousePricePrediction.csv', index_col='Id')
```

```python
df.head()
```

```python
# Separate dependent & independent features
x = df.drop('SalePrice', axis=1)
y = df.SalePrice
```

## <font color='green'>Feature Names</font>

```python
from codeless.impute import cat, num
```

```python
numerical_features = num.GetFeatures(x)
categorical_features = cat.GetFeatures(x)
```

# <font color='red'>num module</font>

```python
# Null values
x[numerical_features].isnull().sum()
```

```python
from codeless.impute import num
```

```python
print(num.__doc__)
```

```python
num.__all__
```

### <font color='green'>Impute Median</font>

```python
# to print the details of the method

print(num.ImputeMmm.__doc__)
```

```python
temp = x.copy()
temp = num.ImputeMmm(temp, numerical_features, strategy='Median')
```

### <font color='green'>Inpute Mean</font>

```python
temp = x.copy()
temp = num.ImputeMmm(temp, numerical_features, strategy='Mean')
```

### <font color='green'>Inpute Mode</font>

```python
temp = x.copy()
temp = num.ImputeMmm(temp, numerical_features, strategy='Mode')
```

### <font color='green'>Capturing NaN values with a new feature</font>

```python
temp = x.copy()
temp = num.ImputeColumn(temp, numerical_features, strategy='Median')
```

### <font color='green'>Random Sample Imputation</font>

```python
temp = x.copy()
temp = num.ImputeSample(temp, numerical_features)
```

### <font color='green'>End Of Distribution</font>

```python
temp = x.copy()
temp = num.ImputeEod(df, numerical_features)
```

```python
df = num.ImputeMmm(df, numerical_features, strategy='Median')
```

# <font color='red'>cat module</font>

```python
from codeless.impute import cat
```

### <font color='green'>Frequent Category Imputation</font>

```python
temp = x.copy()
temp = cat.ImputeMode(temp, categorical_features)
```

### <font color='green'>Replace with Nan Categories</font>

```python
temp = x.copy()
temp = cat.ImputeMissing(temp, categorical_features)
```

### <font color='green'>Capturing NaN values with a new column</font>

```python
temp = x.copy()
temp = cat.ImputeColumn(temp, categorical_features)
```

```python
df = cat.ImputeMissing(df, categorical_features)
```
